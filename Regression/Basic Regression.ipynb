```
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
```

```
#Plots the individual points
plt.scatter([3,-4],[4,5])
```

```
#Plots a straight line given the points
#(3,4) and (-4,5)
plt.plot([3,-4],[4,5])
```

```
#Create the x (features) and y (labels)
x, y = make_regression(n_samples=100, n_features=1, noise=13)

#Create a scatter plot using values of x and y
plt.scatter(x,y)
plt.show()
```

```
for i in range(10):
  print(f'{x[i][0]}--->{y[i]}')
```

```
#Call the LinearRegression Class
reg = LinearRegression()

#Fit the model to the x and y values
#This is where training happens
reg.fit(x, y)

#Using the model, predict the values of y given x
y_pred = reg.predict(x)

#Plot all the real points of (x,y)
plt.scatter(x,y)

#Plot the line using the values (x,y_pred)
plt.plot(x,y_pred)
plt.show()
```

```
x = np.array([[2,5,2],
             [7,2,4],
             [3,6,1],
             [5,3,7]])
```

```
y = np.array([33,35,40,37])
```

```
model = LinearRegression()
model.fit(x,y)
```
